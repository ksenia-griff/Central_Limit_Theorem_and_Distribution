**Distribution Combination**

```python
import matplotlib.pyplot as plt # to chart
import numpy as np#random generator
```

```python
simsize = 1000
#Create arrays of sample size
x = np.random.uniform(1,6,simsize) 
y = np.random.uniform(1,6,simsize)
z= x + y

```

```python
count, bins, ignored = plt.hist(x, 15, normed=True)
plt.xlabel('x')
plt.ylabel('Probability')
histx = plt
histx.show()

count, bins, ignored = plt.hist(y, 15, normed=True)
plt.xlabel('y')
plt.ylabel('Probability')
histy = plt
histy.show()

count, bins, ignored = plt.hist(z, 15, normed=True)
plt.xlabel('z')
plt.ylabel('Probability')
histz = plt
histz.show()

```

```python
plt.figure(1)
plt.subplot(1,3,1)
count, bins, ignored = plt.hist(x, 15, normed=True)
plt.xlabel('x')
plt.ylabel('Probability')
plt.subplot(1,3,2)
count, bins, ignored = plt.hist(y, 15, normed=True)
plt.xlabel('y')
plt.ylabel('Probability')
plt.subplot(1,3,3)
count, bins, ignored = plt.hist(z, 15, normed=True)
plt.xlabel('z')
plt.ylabel('Probability')
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=3.00, hspace=0.50,
                    wspace=0.35)
plt.show()
```

```python
print("mean x:", np.mean(x))
print("mean y:", np.mean(y))
print("mean z:", np.mean(z))
```

**Central Limit Theorem**

```python
from numpy import ndarray
samplesize = 100
samples = 10000
xmeandist = ndarray((samples,),float)
ymeandist = ndarray((samples,),float)
zmeandist = ndarray((samples,),float)
```

```python
for i in range(0,samples):
    xsample = np.random.choice(x, size=samplesize, replace=True, p=None)
    xmeandist[i]= np.mean(xsample)
    
    ysample = np.random.choice(y, size=samplesize, replace=True, p=None)
    ymeandist[i]= np.mean(ysample)
    
    zsample = np.random.choice(z, size=samplesize, replace=True, p=None)
    zmeandist[i]= np.mean(zsample)
```

```python
plt.figure(1)
plt.subplot(1,3,1)
count, bins, ignored = plt.hist(xmeandist, 15, normed=True)
plt.xlabel('x means')
plt.ylabel('Probability')

plt.subplot(1,3,2)
count, bins, ignored = plt.hist(ymeandist, 15, normed=True)
plt.xlabel('y means')
plt.ylabel('Probability')

plt.subplot(1,3,3)
count, bins, ignored = plt.hist(zmeandist, 15, normed=True)
plt.xlabel('z means')
plt.ylabel('Probability')
           
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=3.00, hspace=0.50,
                    wspace=0.35)           
plt.show()
```
